```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from imblearn.combine import SMOTEENN

pd.pandas.set_option('display.max_columns' , None)
```

```python
df = pd.read_csv('df_churn.csv')
df.head(5)
```

```python
df = df.drop('Unnamed: 0' ,axis = 1)
df.head(3)
```

```python
#seperating dependent and independent variable 

x = df.drop('Churn' , axis = 1)
y = df['Churn']
```

```python
x_train , x_test , y_train , y_test = train_test_split( x , y  , test_size = 0.2 , random_state = 42)
```

### Decision Tree Classifier

```python
dt = DecisionTreeClassifier(criterion = 'gini' , random_state = 100 , max_depth = 6 , min_samples_leaf = 8)
dt.fit(x_train , y_train)
```

```python
y_pred = dt.predict(x_test)
y_pred
```

```python
#comparing the actual value with predicted value

print(classification_report(y_test , y_pred  , labels = [0 , 1]))
```

```python
print(confusion_matrix(y_test , y_pred))
```

We can see that the score is very low for churner - 1 . The dataset is imbalanced .

```python
#since the dataset is imbalanced we will be using SMOTE-ENN technique .

sm = SMOTEENN()
x_resampled , y_resampled = sm.fit_resample(x,y)
```

```python
xr_train , xr_test , yr_train , yr_test = train_test_split( x_resampled , y_resampled , test_size = 0.2 , random_state = 43)
```

```python
dt_smote = DecisionTreeClassifier(criterion = 'gini' , random_state = 100 , max_depth = 6 , min_samples_leaf = 8)
dt_smote.fit(xr_train , yr_train)
```

```python
ypred_smote  = dt_smote.predict(xr_test)
```

```python
#comparing the actual value with predicted value

print(classification_report(yr_test , ypred_smote  , labels = [0 , 1]))
```

```python
print(confusion_matrix(yr_test , ypred_smote))
```

WE  HAVE 91% ACCURACY FOR THIS MODEL .
